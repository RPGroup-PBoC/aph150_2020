### 1. Microplastic Mass Concentration Dataset

```python
concentrations = pd.read_csv(data_dir + "Microplastic_Numerical_and_Mass_Concentration.csv")
concentrations["Datetime PST"] = pd.to_datetime(concentrations["Datetime PST"])
print(concentrations.columns)
```

```python
concentrations.head()
```

```python
year_sorted = concentrations.groupby(concentrations['Datetime PST'].map(lambda x: x.year))
years = np.sort(concentrations['Datetime PST'].map(lambda x: x.year).unique())
```

```python
# Get a running tally of abundance values 
yearly_abundance = year_sorted.agg("mean")
abundance_ratio = yearly_abundance["Microplastic Abundance"]
microplastic_mass = yearly_abundance["Microplastic Mass (mg/m³)"]
```

```python
yearly_abundance.head(10)
```

```python
plt.semilogy(years, abundance_ratio)
plt.xlabel("Years")
plt.ylabel("Microplastic Abundance")
plt.title("Microplastic Abundance Over Time")
plt.show()
```

```python
concentration_sorted = concentrations.sort_values(by=["Datetime PST"])
```

```python
concentration_sorted.head()
```

```python
times = concentration_sorted["Datetime PST"]
abundance = concentration_sorted["Microplastic Abundance"]
```

```python
plt.plot(times, abundance)
plt.xlabel("Time")
plt.ylabel("Microplastic Abundance Ratio")
plt.title("Microplastic Abundance Over Time")
plt.show()
```

```python
# Plot Latitude and Longitude for Given Year 
concentrations["Year"] = concentrations["Datetime PST"].map(lambda x: x.year)
for year in years:
    year_coords = concentrations[concentrations["Year"] == year][["Latitude (º)", "Longitude (º)"]]
    plt.scatter(year_coords["Longitude (º)"], year_coords["Latitude (º)"], label=year, s=25)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geolocations of Microplastic Sampling Locations")
plt.legend()
plt.show()
```

```python
# Need Basemap installation to see this
# https://stackoverflow.com/questions/42299352/installing-basemap-on-mac-python
from mpl_toolkits.basemap import Basemap
```

```python
lat0 = np.mean(concentrations["Latitude (º)"])
lon0 = np.mean(concentrations["Longitude (º)"])

fig = plt.figure(figsize=(8, 8))


m = Basemap(projection='lcc', resolution=None,
            width=15E6, height=15E6, 
            lat_0=lat0, lon_0=lon0,)
m.etopo(scale=0.25, alpha=0.5)

# Map (long, lat) to (x, y) for plotting
xs = []
ys = []
for year in years:
    year_coords = concentrations[concentrations["Year"] == year][["Latitude (º)", "Longitude (º)"]]
    
    year_latitudes = list(year_coords["Latitude (º)"])
    year_longitudes = list(year_coords["Longitude (º)"])
    
    
    x,y = m(year_longitudes, year_latitudes)
    plt.scatter(x, y, label=year)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geolocations of Microplastic Sampling Locations")
plt.legend()
plt.show()
```

```python
# Look at the relative abuncance 
n_pacific_years = [np.mean(years[[0, 1]]), np.mean(years[[6, 7]])]
n_pacific_masses = [np.mean(microplastic_mass.values[[0, 1]]), np.mean(microplastic_mass.values[[6, 7]])]
plt.bar(n_pacific_years, n_pacific_masses, width=20)
plt.title("Microplastic Abundance 1970s vs 2010")
plt.xlabel("Years")
plt.ylabel("Microplastic Mass (mg/m³)")
print("Microplastic Mass Ratio 2010 to 1970: %0.2f" % (n_pacific_masses[1] / n_pacific_masses[0]))
```

### 2. Marine Pollution Estimation Dataset 

```python
marine_pollution = pd.read_excel(data_dir + "PlasticMarinePollutionGlobalDataset.xlsx", header=[0],skiprows=0, parsedates=["Date"], converters={'Date': str})
```

```python
marine_pollution = marine_pollution.drop(index=0)
marine_pollution.head()
```

```python
list(marine_pollution.columns)
```

```python
def undate(x):
#     if pd.isnull(x):
#         return x
    try:
        return x.strftime('%d/%m/%Y')
    except AttributeError:
        return x
    except Exception:
        raise

marine_pollution["Date"] = marine_pollution["Date"].apply(undate)
marine_pollution["Date"] = pd.to_datetime(marine_pollution["Date"])
marine_pollution["CD1  (/km^2)"] = pd.to_numeric(marine_pollution["CD1  (/km^2)"])
marine_pollution["CD2  (/km^2)"] = pd.to_numeric(marine_pollution["CD2  (/km^2)"])
marine_pollution["CD3  (/km^2)"] = pd.to_numeric(marine_pollution["CD3  (/km^2)"])
marine_pollution["CD4  (/km^2)"] = pd.to_numeric(marine_pollution["CD4  (/km^2)"])
marine_pollution["WD1 (g/km^2)"] = pd.to_numeric(marine_pollution["WD1 (g/km^2)"])
marine_pollution["WD2 (g/km^2)"] = pd.to_numeric(marine_pollution["WD2 (g/km^2)"])
marine_pollution["WD3 (g/km^2)"] = pd.to_numeric(marine_pollution["WD3 (g/km^2)"])
marine_pollution["WD4 (g/km^2)"] = pd.to_numeric(marine_pollution["WD4 (g/km^2)"])
```

```python
pollution_sorted = marine_pollution.sort_values(by=["Date"])
pollution_years = np.sort(pollution_sorted['Date'].map(lambda x: x.year).unique())
marine_pollution["Year"] = marine_pollution['Date'].map(lambda x: x.year)
pollution_year_grouping = pollution_sorted.groupby(pollution_sorted['Date'].map(lambda x: x.year))
yearly_pollution = pollution_year_grouping.agg("mean")
```

```python
pollution_years
```

```python
yearly_pollution.head()
```

```python
lat0 = np.mean(marine_pollution["Latitude"])
lon0 = np.mean(marine_pollution["Longitude"])

fig = plt.figure(figsize=(16, 8))


m = Basemap(projection='lcc', resolution=None,
            width=3.5E7, height=1.4E7, 
            lat_0=lat0, lon_0=lon0,)
m.etopo(scale=0.25, alpha=0.5)

# Map (long, lat) to (x, y) for plotting
xs = []
ys = []
for year in pollution_years:
    year_coords = marine_pollution[marine_pollution["Year"] == year][["Latitude", "Longitude"]]
    
    year_latitudes = list(year_coords["Latitude"])
    year_longitudes = list(year_coords["Longitude"])
    
    
    x,y = m(year_longitudes, year_latitudes)
    plt.scatter(x, y, label=year, s=20)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geolocations of Microplastic Sampling Locations")
plt.legend()
plt.show()
```
